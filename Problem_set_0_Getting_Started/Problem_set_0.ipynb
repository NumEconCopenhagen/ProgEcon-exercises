## Task 2: Understanding code

For each snippet below, **first write down your expected output on paper**. Then run the cell and check whether you were correct.

**indexing and references**

```python
x = [1,2,3]
y = x
y[-1] = 4
print(x)
```

**slicing**

```python
x = [0,1,2,3,4,5]
print(x[:2])
print(x[2:])
```

**loops - break**

```python
for i in range(5):
    if i >= 2: break
    print(i)
```

**loops - continue**

```python
for i in range(5):
    if i == 2: continue
    print(i)
```

**conditionals**

```python
x = 3
if x > 3:
    print('too big')
elif x < 1:
    print('too small')
else:
    print('just right')
```

**functions and scope**

```python
a = 1
def f(x):
    return x+a # a is global variable
def g(x,a=1):
    return x+a # a is local variable
print(f(1))
print(g(1))
a = 2
print(f(1))
print(g(1))
```

**dictionaries**

```python
d = {'a': 1, 'b': 2}
for key in d:
    d[key] = d[key]*2
print(d['a'] + d['b'])
print('c' in d)
```

**list comprehension and zip**

```python
a = [1,2,3]
b = [4,5,6]
c = [i+j for i,j in zip(a,b)]
print(c)
```

**floating point numbers**

```python
print(0.1 + 0.2 == 0.3)
import math
print(math.isclose(0.1 + 0.2, 0.3))
```

**classes**

```python
class SquareClass:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def size(self):
        return self.length*self.width
square = SquareClass(2,2)
print(square.size())
```

**numpy - view and copies I**

```python
import numpy as np
x = np.array([1,2,3])
y = x
x += 1
print(y)
x[:] = x + 1
print(y)
x = x + 1
print(y)
```

**numpy - view and copies II**

```python
import numpy as np
A = np.array([0,1,2,3,4])
B = A[1:4]
B[:] = 0
print(A)
```

**numpy - broadcasting**

```python
import numpy as np
A = np.array([[1,2,3],[4,5,6]])
b = np.array([10,20,30])
print(A + b)
```
